```python
import numpy as np 
import matplotlib.pyplot as mt
x=[20,25,28,32,35,40,42]
d=np.arange(len(x))
width=0.25
mt.xticks(ticks=d,labels=x)
m_y1=[1920,2000,2241,2320,2457,2820,2975]
mt.bar(d,m_y1,width=width,label="mol",color="red")
m_y2=[1500,1580,1741,1820,1857,2000,2075]
mt.bar(d+width,m_y2,label="dok",width=width)
mt.legend()
mt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as mt
x=[20,25,28,32,35,40,42]
d=np.arange(len(x))
width=0.25
mt.yticks(ticks=d,labels=x)
m_y1=[1920,2000,2241,2320,2457,2820,2975]
mt.barh(d,m_y1,label="mol",color="red")
mt.legend()
mt.show()
```
