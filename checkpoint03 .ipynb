```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(data)
```

```python
this heatmap replaces each of the variables on the features with a list of numeric variables in the dataset.
Each cell depicts the relationship between the intersecting variables, such as a linear correlation,If the value between 
the two variables is less than one, then thy have a negative correlation and if it was positive than the 
correlation is positive.
```

```python
cleanup={"Sex":{"male":1,"female": 0}} 
data.replace(cleanup, inplace=True)
data.groupby(by="title").agg(Survived_rates=("Survived","mean")).plot(kind='bar') 
data.groupby(by="title").agg(age_rates=("Age","mean")).plot(kind='bar')   
data.groupby(by="title").agg(Fare_payed_rates=("Fare","mean")).plot(kind='bar')    
data.groupby(by="title").agg(Sex_rates=("Sex","mean")).plot(kind='bar')
```
